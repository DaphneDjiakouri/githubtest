```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

**1.** Plot the function $f(x) = x^3 − 2x + 2$ within the range x ∈ [−2, 2].

```python
# _Function to plot
X = np.arange(-2, 2, 0.25)
Y = X**3 - 2*X + 2;

# Plot
line, = plt.plot(X, Y)
plt.show()
```

**2.** We compute analytically the points x* that satisfy $f'(x) = 0$

\begin{align*}
f'(x) &= 0 \\
\Rightarrow 3x^2 - 2 &= 0 \\
\Rightarrow x^2 &= \frac{2}{3} \\
\Rightarrow x &= \pm \sqrt{\frac{2}{3}}
\end{align*}

The obtained result is congruent with the plot of the previous point, as we can observe from the plot that there is a maximum between -1.0 and-0.5 and a minimun between 0.5 and 1.0.

**3.** We are now going to check which of the latter points x* are minima (or a maxima). For that purpose we will use a 2nd order Taylor expansion around point x* :
$f(x^* + d) \approx f(x) + df'(x) + \frac{1}{2}d^2f''(x^*)$

In order for x* to be a minimum we need $f''(x)$ to be positive and in order for x* to be maximum we need $f''(x)$ to be negative.

We evaluate $f''(x)$ at the points of interest:

$f'(x) = 3x^2 - 2$

$f''(x) = 6x$

$f''(\sqrt{2/3}) > 0%$ So, the point $x =\sqrt{2/3}$ is a minimum.

$f''(-\sqrt{2/3}) < 0$ So, the point $x =-\sqrt{2/3}$ is a maximum.


**4.** We plot $f''(x)$ within the range x ∈ [−2, 2].

```python
# _Function to plot
X = np.arange(-2, 2, 0.25)
Y = 6*X;

# Plot
line, = plt.plot(X, Y)
plt.show()
```


When x ∈ (-2, 0), $f''(x)$ is negative, so our original function $f(x)$ is concave.

When x ∈ (0, -2), $f''(x)$ is positive, so our original function $f(x)$ is convex.

**1.** Plot the function $f(x) = x_1^2 + x_2^2$

```python
# We define the given function and plot it as we did before
def f(x_1, x_2):
    return x_1**2 + x_2**2
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
x_1 = np.arange(-5, 5, 0.25)
x_2 = np.arange(-5, 5, 0.25)
x_1, x_2 = np.meshgrid(x_1, x_2)
z = f(x_1, x_2)

surf = ax.plot_surface(x_1, x_2, z)
plt.show()
```

**2.** We will analytically compute the gradient of the function, $∇f(x)$, and compute the point $x^*$ at which $∇f(x^*) = 0$.


$\nabla f(x) = \left(\frac{\delta f(x)}{\delta x_1}, \frac{\delta f(x)}{\delta x_2}\right)^T = \left(2x_1, 2x_2\right)^T$

$\nabla f(x) = 0 \quad \Rightarrow \quad (2x_1, 2x_2)^T = (0, 0)^T \quad \Rightarrow \quad (x_1, x_2)^T = (0, 0)^T$





**3.** We will check now if $x^*$ is a minimum or a maximum,by using the Taylor expansion, up to second order and we compute the Hessian matrix, $∇^2f(x)$, at the point $x = x^*$. We obtain
$\nabla^2 f(x^*) = \begin{pmatrix} 2 & 0 \\ 0 & 2 \end{pmatrix}$

For a higher dimensional problem we are sure that the quadratic
approximation is convex and that we have a minimum if

$d^T \nabla^2f(x^*) d > 0, \quad d \neq 0 \quad (3)$ for $d \in \mathbb{R}^2$

We have a maximum if

$d^T \nabla^2f(x^*) d < 0, \quad d \neq 0 \quad (4)$ for $d \in \mathbb{R}^2$

The previous conditions can be verified by computing the eigenvalues of $∇^2f(x^∗)$. If all eigenvalues are strictly positive, equation (3) is satisfied. If all eigenvalues are strictly negative, equation (4) is satisfied.

To calculate the eigenvalues, we will find the solutions of the characteristic polynomial of $∇^2f(x)$:

$\left| \begin{matrix} 2 - \lambda & 0 \\ 0 & 2 - \lambda \end{matrix} \right| = 0 \Rightarrow (2 - \lambda)^2 = 0 \Rightarrow \lambda = 2$

Both eigenvalues are strictly positive so we have a minimum at (0,0).

**1.** We wiil plot the following function within the range x1 ∈ [−2, 2] and x2 ∈ [−1, 1],
 $f(x_1, x_2) = x_1^2 \left(4 - 2.1x_1^2 + \frac{1}{3}x_1^4\right) + x_1x_2 + x_2^2 \left(-4 + 4x_2^2\right)$
 and observe where the minima (and maxima) may be.


```python
def f(x_1, x_2):
    return x_1**2*(4 - 2.1*x_1**2 + 1/3*x_1**4) + x_1*x_2 + x_2**2*(-4 + 4*x_2**2)

fig = plt.figure()
ax = plt.axes(projection='3d')
x_1 = np.arange(-2, 2, 0.1)
x_2 = np.arange(-1, 1, 0.1)
x_1, x_2 = np.meshgrid(x_1, x_2)
z = f(x_1, x_2)

surf = ax.plot_surface(x_1, x_2, z)
plt.show()
```

There seem to be a local minimum in each corner of the plot. More specific, there is a local maximum close to $(x_1,x_2)=(0,-1)$ and a local minimum at $(x_1,x_2)=(-0.5,0)$.

**2.** We will analytically compute the gradient $∇f(x)$:

$\nabla f(x) = (8x_1 - 8.4x_1^3 + 2x_1^5 + x_2, x_1 - 8x_2 + 16x_2^3)$



**3.** Numerically we will compute an approximation of the points $x^*$ at which $\nabla f(x^*)=0$.


(a) Let's evaluate $\|\nabla f(x)\|^2$
at the previous range.

```python
import sympy as sp
import numpy as np
```

```python
# First we define the function and calculate the gradient vector of the function

x_1, x_2 = sp.symbols('x_1 x_2')
f =  x_1**2*(4 - 2.1*x_1**2 + 1/3*x_1**4) + x_1*x_2 + x_2**2*(-4 + 4*x_2**2)

grad_f = [sp.diff(f, var) for var in (x_1, x_2)]
print('Gradient of the function:', grad_f)

```

```python

def eval_grad_norm(grad_f, c_1, c_2):
  '''
  This function needs as input the gradient vector of the function and the two coordinates for which we want to compute the square norm
  '''
  # first we evaluate the gradient at that point
  valor_grad_f = np.array([grad.subs({x_1: c_1, x_2: c_2}) for grad in grad_f])
  # print(valor_grad_f)

  # we compute the norm
  norm = np.sum(valor_grad_f**2)

  # print(norm)
  # print(valor_grad_f[0]**2 + valor_grad_f[1]**2)

  return norm

eval_grad_norm(grad_f, -1, -2)
```

```python
x_1_values = np.arange(-2, 2, 0.005)
x_2_values = np.arange(-1, 1, 0.005)

A = np.zeros((len(x_1_values), len(x_2_values)))
```

```python
for i in range(len(x_1_values)):
  for j in range(len(x_2_values)):
    A[i, j] = eval_grad_norm(grad_f, x_1_values[i], x_2_values[j])

A
```
